```python
url="https://raw.githubusercontent.com/dineshpiyasamara/LaptopPricePredictor/master/model%20building/laptop_price.csv"
data=pd.read_csv(url,encoding='latin-1')
data.head()
```

```python
#remove GB of ram value
data['Ram']= data['Ram'].str.replace('GB','').astype('int32')
```

```python
#remove  kg  of weight value
data['Weight']= data['Weight'].str.replace('kg','').astype('float32')
```

```python
#increase categories coz one hot encording
def add_company(inp):
   
    
    # Define a set of company names to check against
    other_companies = {'Samsung', 'Razer', 'Mediacom', 'LG', 'Microsoft', 'Xiaomi', 'Vero', 'Chuwi', 'Google', 'Fujitsu', 'Huawei'}
    
    # Check if the input is in the set of other companies
    if inp in other_companies:
        return 'Other'
    else:
        return inp

# Apply the function to the 'Company' column
data['Company'] = data['Company'].apply(add_company)
```

```python
data['Touchscreen']=data['ScreenResolution'].apply(lambda x:1 if 'TouchScreen' in x else 0)
data['Ips']=data['ScreenResolution'].apply(lambda x:1 if 'TouchScreen' in x else 0)
```

```python

data['Cpu'].value_counts()
len(data['Cpu'].value_counts())
```

```python
data['cpu_name']=data['Cpu'].apply(lambda x: " ".join(x.split()[0:3]))
```

```python
data['cpu_name']
```

```python
def set_processor(name):
    if 'Intel Core i7' in name or 'Intel Core i5' in name or 'Intel Core i3' in name:
        return name
    elif name.split()[0] == 'AMD':
        return 'AMD'
    else:
        return 'Other'



# Apply the function to the 'cpu_name' column
data['cpu_name'] = data['cpu_name'].apply(set_processor)


```

```python
data['gpu_name']=data['Gpu'].apply(lambda x: " ".join(x.split()[0:1]))
```

```python
data=data[data['gpu_name']!='ARM']
```

```python
data['OpSys'].value_counts()
```

```python
# Define the function for categorization
def categorize_os(os_name):
    if os_name == 'Windows 10' or  os_name == 'Windows 7' or  os_name == 'Windows 10 S':
        return 'Windows'
    elif  os_name == 'Linux':
        return 'Linux'
    elif  os_name == 'macOS' or  os_name == 'Mac OS X':
        return 'Mac'
    else:
        return 'Other'

# Apply the function to the 'OpSys' column
data['OpSys'] = data['OpSys'].apply(categorize_os)
```

```python
data['OpSys'].value_counts()
```

```python
data.head(2)
```

```python
data = data.drop(columns=['laptop_ID','Inches','Product','ScreenResolution','Cpu','Gpu'])
```

```python
data.head(2)
```

```python
data = pd.DataFrame(data)

# Columns to one-hot encode
columns_to_encode = ['Company','TypeName', 'OpSys', 'cpu_name', 'gpu_name']

# One-hot encoding
data = pd.get_dummies(data, columns=columns_to_encode)

# Print the DataFrame to see the initial encoding
print("Initial encoded DataFrame:\n", data)

# Identify boolean columns
bool_cols = data.select_dtypes(include='bool').columns

# Convert only boolean columns to int
data[bool_cols] = data[bool_cols].astype(int)
```

```python
x = data.drop('Price_euros',axis=1)# remove price from features
y=data['Price_euros']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
def model_accuracy(model):
    model.fit(x_train,y_train)
    accuracy=model.score(x_test,y_test)
    print(str(model)+'------->'+str(accuracy))
```

```python
from sklearn.linear_model import LinearRegression #linear regression
lr=LinearRegression()
model_accuracy(lr)

from sklearn.tree import DecisionTreeRegressor #DecisionTreeRegressor
dt=DecisionTreeRegressor()
model_accuracy(dt)

from sklearn.ensemble import RandomForestRegressor #linear regression
rf=RandomForestRegressor()
model_accuracy(rf)


```

### According to accuracy we choose  RandomForestRegressor to train model

```python
 #hyer parameters tuning for increase accuracy
from sklearn.model_selection import GridSearchCV

parameters ={

    'n_estimators':[10,50,100],
     'criterion' :['squared_error','absolute_error','poisson']
}

grid_obj=GridSearchCV(estimator=rf,param_grid=parameters)
grid_fit =grid_obj.fit(x_train,y_train)
best_model =grid_fit.best_estimator_
best_model
```

```python
best_model.score(x_test,y_test)
```
